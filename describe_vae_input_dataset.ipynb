```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.colors as colors
from matplotlib.colors import LogNorm
import os
import h5py
```

```python
def plot_hist_2d( x, y, xlabel, ylabel, title):
    fig = plt.figure()
    plot_hist_2d_on_axis( plt.gca(), x, y, xlabel, ylabel, title )
    plt.tight_layout()
    #fig.savefig(CONFIG['plotdir'] + plotname + '.png')
    #plt.close()
    plt.draw()

def plot_hist_2d_on_axis( ax, x, y, xlabel, ylabel, title ):
    ax.hist2d(x, y, bins=100, norm=colors.LogNorm())
    ax.set_xlabel(xlabel)
    ax.set_ylabel(ylabel)
    ax.set_title(title)

def get_dijet_features_from_file( file_path ):
    print('reading ', input_file)
    ff = h5py.File(input_file, 'r')
    keys = list(ff.keys())
    print(ff, ': ', keys)
    return np.array(ff.get('eventFeatures'))
```

```python
plot_hist_2d(evts[:,0],evts[:,9],'mass_jj','delta eta jj', 'mass jj vs delta eta QCD signal')
plot_hist_2d(evts[:,0],evts[:,1],'mass_jj','pt j1', 'mass jj vs pt j1 QCD signal')
plot_hist_2d(evts[:,9],evts[:,1],'delta eta jj','pt j1', 'delta eta jj vs pt j1 QCD signal')
```
