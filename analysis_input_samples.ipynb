```python
import os
from collections import OrderedDict
```

```python
import data_sample as ds
import string_constants_vae_check as sc_check
import plotting_util as pu
```

```python
def plot_feature( sample_dict, feature_name, sample_names=None ):
    if not sample_names: 
        sample_names = sample_dict.keys()
    legend = [ sample_dict[s].title() for s in sample_names ]
    feature = [ sample_dict[s][feature_name] for s in sample_names ]
    _ = pu.plot_hist(feature,legend=legend,xlabel=feature_name,title=r'distribution '+feature_name,legend_loc=(1.05,0))
```
