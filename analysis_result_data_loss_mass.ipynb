```python
import os
```

```python
import reading_util as ru
import writing_util as wu
import plotting_util as pu 
import string_constants as sc
import data_sample as ds
import parameter as pa
```

# Datasets

```python
SM_samples = ['qcdSigAll']
BSM_samples = ['GtoWW15na', 'GtoWW30na', 'GtoWW45na']
all_samples = SM_samples + BSM_samples
```

```python
# read in data
input_data = {}
for name in all_samples:
    print(name)
    path = os.path.join( sc.concat_result_dir, sc.sample_loc[name] + sc.concat_result_suffix )
    input_data[name] = ds.read_datasample_from_file( name, os.path.join( sc.result_dir, path ) )
```

## plot mass ( mjj, mj1, mj2 ) for all datasets

## mjj

```python
masses = [sample['mJJ'] for sample in input_data.values()]
labels = [sample.title() for sample in input_data.values()]

_ = pu.plot_hist(masses, xlabel=r'$m_{JJ}$', legend=labels, title=r'$m_{JJ}$ distribution',ylogscale=False)
```

## m j1

```python
masses = [sample['j1M'] for sample in input_data.values()]

_ = pu.plot_hist(masses, xlabel=r'$m_{J1}$', legend=labels, title=r'jet 1 mass distribution',ylogscale=False)
```

## m j2

```python
masses = [sample['j2M'] for sample in input_data.values()]

_ = pu.plot_hist(masses, xlabel=r'$m_{J2}$', legend=labels, title=r'jet 2 mass distribution',ylogscale=False)
```

# mass vs loss L1 & L2

```python
loss_combi = pa.loss_strategy_dict['s5'] # s5 = L1 & L2
```

## mjj vs l1 & l2

```python
for name in all_samples:
    pu.plot_hist_2d( input_data[name]['mJJ'], loss_combi(input_data[name]), xlabel=r'$m_{jj}$', ylabel=r'$L_1$ \& $L_2$', title=r'$m_{JJ}$ vs loss ' + input_data[name].title())
```

## m j1  vs l1 & l2

```python
for name in all_samples:
    pu.plot_hist_2d( input_data[name]['j1M'], loss_combi(input_data[name]), xlabel=r'$j1 m$', ylabel=r'$L_1$ \& $L_2$', title=r'jet 1mass vs loss ' + input_data[name].title())
```
